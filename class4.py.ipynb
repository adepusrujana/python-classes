```python
Here, the \t is replaced with enough spaces to align to the next multiple of 4 columns.

How it works:

Default behavior: If no argument is passed, each tab is expanded to align at every 8th column.

Custom tab size: Passing an integer changes the alignment width. For example, expandtabs(4) aligns tabs at every 4th column.

Column-based replacement: The number of spaces inserted depends on the current position in the string, not just a fixed replacement count.

More Examples:

Default tab size (8 spaces)

s = "A\tB\tC"
print(s.expandtabs())
Copy
Output:

A B C
Copy
Custom tab size

s = "Python\tis\tfun"
print(s.expandtabs(4))
Copy
Output:

Python is fun
Copy
Mixed text and tabs

s = "A\tBC\tDEFG"
print(s.expandtabs(5))
Copy
Output:

A BC DEFG
```
