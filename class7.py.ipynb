```python
lis=[1,2,3,[5,3]]
lis1=[]
#flatten the list
for i in lis:
    if type(i)==list:
        lis1.extend(i)
    else:
        lis1.append(i)
print(lis1)
```

```python
n=[1,2,3,4,5]
sum_=0
for i in range(len(n)):
    sum_+=n[i]
print(sum_)

```

```python
n=[1,2,3,4,5]
i=0
sum_=0
while i<len(n):
    sum_=sum_+n[i]
    i+=1
print(sum_)

```

```python
# max element
arr=[3,5,6,2]
##using while
i=0
max_=0
while i<len(arr):
    if arr[i]>max_:
        max_=arr[i]
    i+=1
print(max_)
## using for
max_=0
for i in range(len(arr)):
    if arr[i]>max_:
        max_=arr[i]
print(max_)

```
